# 카테고리로 분류

```python
df = pd.read_csv('C:/Users/이재진/Documents/Codeit/python_data_science/data/laptops.csv')
df.head()
```

```python
brand_nation = {
    'Dell': 'U.S',
    'Apple': 'U.S',
    'Acer': 'Taiwan',
    'HP': 'U.S',
    'Lenovo': 'China',
    'Alienware': 'U.S',
    'Microsoft': 'U.S',
    'Asus': 'Taiwan'
}
```

```python
df['brand'].map(brand_nation)
```

```python
df['brand_nation'] = df['brand'].map(brand_nation)
```

```python
df
```

# group by

```python
nation_groups = df.groupby('brand_nation')
type(nation_groups)
```

```python
nation_groups.count()
```

```python
nation_groups.max()
```

```python
nation_groups.mean()
```

```python
nation_groups.first()
```

```python
nation_groups.last()
```

```python
nation_groups.plot(kind='box', y='price')
```

```python
nation_groups.plot(kind='hist', y='price')
```

# 데이터 합치기

```python
price_df = pd.read_csv('C:/Users/이재진/Documents/Codeit/python_data_science/data/vegetable_price.csv')
quantity_df = pd.read_csv('C:/Users/이재진/Documents/Codeit/python_data_science/data/vegetable_quantity.csv')
```

```python
price_df
```

```python
quantity_df
```

```python
pd.merge(price_df,quantity_df, on ='Product')
```

```python
pd.merge(price_df,quantity_df, on ='Product', how='left')
```

```python
pd.merge(price_df,quantity_df, on ='Product', how='right')
```

```python
pd.merge(price_df,quantity_df, on ='Product', how='outer')
```
